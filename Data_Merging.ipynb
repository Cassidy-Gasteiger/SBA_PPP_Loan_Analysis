```python
ppp = pd.read_csv(
    open(
        r'C:\Users\cmgas\OneDrive\Business_Practicum\Data\GeorgiaSBAPPPDataAll_060921.csv', 'rb'),
)
```

```python
ppp.head(5)
```

```python
agg_ppp = ppp[['CountyID', 'InitialApprovalAmount']]
agg_ppp.head(5)
```

```python
loan_amt_by_county = agg_ppp.groupby('CountyID').sum()
loan_amt_by_county = loan_amt_by_county.reset_index()
loan_amt_by_county = loan_amt_by_county.rename(columns={'InitialApprovalAmount': 'Total Amt of Loans Received'})
loan_amt_by_county.head(5)
```

```python
loan_no_by_county = agg_ppp.groupby('CountyID').count()
loan_no_by_county = loan_no_by_county.reset_index()
loan_no_by_county = loan_no_by_county.rename(columns={'InitialApprovalAmount': 'Number of Loans Received'})
loan_no_by_county
```

```python
loan_no_by_county['CountyID'] = '1' + loan_no_by_county['CountyID'].astype(str)
loan_no_by_county['CountyID'] = loan_no_by_county['CountyID'].str.rstrip('.0')


loan_amt_by_county['CountyID'] = '1' + loan_amt_by_county['CountyID'].astype(str)
loan_amt_by_county['CountyID'] = loan_amt_by_county['CountyID'].str.rstrip('.0')
```

```python
loan_no_by_county
```

```python
loan_no_final = pd.merge(merge_7, loan_no_by_county, on = ['CountyID'])
ppp_dataframe = pd.merge(loan_no_final, loan_amt_by_county, on = ['CountyID'])
ppp_dataframe
```
